```python
from typing import TypedDict

class User(TypedDict):
    name: str
    age: int 

user:User = {
    "name": "James",
    "age": 32
}
```

```python
import json

class Address(TypedDict):
    city: str
    country: str

class Student(TypedDict):
    name:str
    level: int
    address:Address


address:Address = {
    "city": "Lagos",
    "country":"Nigeria"
}
student: Student = {
    "name": "Emmanuel Bakare",
    "level": 3,
    "address": address

}

print("Students :",student)
print("Address :", address)
```

```python
class TestState(TypedDict):
    counter: int 

def get_state(state:TestState):
    if not state.get('counter',0):
        state['counter']=1
        return "There is nothing inside"
    else: 
        return f"content of state {state['counter']}"


```

```python
new_state = get_state({
    "counter":2
})
print(new_state)
```
