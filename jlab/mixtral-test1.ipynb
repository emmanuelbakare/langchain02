```python
from langchain_community.chat_models import ChatOllama
from langchain_core.tools import tool 
from langchain_core.messages import BaseMessage, HumanMessage
from langgraph.graph import START, END, StateGraph

llm = ChatOllama(model="mixtral")

@tool 
def get_weather(city:str):
    """ Get the weather of the city"""

    weather_data = {
        "London": "Sunny 23 F",
        "New York": " Cloudy 34 F",
        "Nigeria": "Rainy 22 F"
    }

    return weather_data.get(city, f"No weather data for {city}")

@tool
def calculate_tip(amount:float, tip_percentage:float):
    """calculate the tip  from the amount paid"""

    return round(amount *(tip_percentage/100),2)


all_tools = [get_weather,calculate_tip]

llm_with_tools = llm.bind_tools(all_tools)

response = llm_with_tools.invoke("I bought a $200 item and gave a $15 tip how much did I give")

for tool_call in response.tool_calls:
    if tool_call['name'] == "calculate_tip":
        answer = calculate_tip.invoke(tool_call['args'])
        result = f"Item bought is ${tool_call['args']['amount']}. Giving a ${tool_call['args']['tip_percentage']} tip costs {answer}"
    elif tool_call['name'] == "get_weather":
        answer = get_weather.invoke(tool_call['args'])
        result = f"Weather for {tool_call['args']['city']} is {answer}"

print(result)
```
